```python
'''
20_9.巢狀迴圈-印圖形1(10.0)
'''

N = int(input())

if N < 1 or N > 9 :
    print("錯誤")
    
    
def main():
    for i in range(1,N+1): # 外層迴圈
        print(str(i)*i)

if __name__ == "__main__":
    main()




```

```python
'''
20_9.巢狀迴圈-印圖形1(10.0)
'''

N = int(input())

if N < 1 or N > 9 :
    print("錯誤")
    
    
def main():
    for i in range(1,N+1): # 外層迴圈
        for j in range(1,i+1): # 內層迴圈
            print(j,end="") # 避免換行
        print() # 換行

if __name__ == "__main__":
    main()




```
