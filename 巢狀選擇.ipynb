寫一個程式讓使用者輸入分數，印出分數等級。

如100 到 90 之間是等級 A，印出Your score is 9x and degree is A!其餘類推如下：
89-80 B
79-70 C 
69-60 D
59-0 F

如果輸入的資料是超過100或者小於0，則顯示資料錯誤(error data)，此程式可以執行2次。

輸入範例: 
85
101

輸出範例:
Your score is 85 and degree is B!
error data

範例程式：
num = int(input())
if num > 100:
    print("error data")
else:
    if num>=90:
        print("Your score is {0} and degree is A!".format(num))
    else:
        if num >= 80:
            print("Your score is {0} and degree is B!".format(num))
        else:
            if num >= 70:
                print("Your score is {0} and degree is C!".format(num))
            else:
                if num >= 60:
                    print("Your score is {0} and degree is D!".format(num))
                else:
                    if num>=0:
                        print("Your score is {0} and degree is F!".format(num))
                    else:
                        print("error data")

```python
'''
14_1.巢狀選擇-分數等第(40.0)
'''

#與範例程式相同意思
for i in range(2):
    score = int(input())
    if score > 100:
        print("error data")
    elif 90 <= score <= 100 :
        print(f"Your score is {score} and degree is A!")
    elif 80 <= score <=89 :
        print(f"Your score is {score} and degree is B!")
    elif 70 <= score <=79 :
        print(f"Your score is {score} and degree is C!")
    elif 60 <= score <=69 :
        print(f"Your score is {score} and degree is D!")
    elif 50 <= score <=59 :
        print(f"Your score is {score} and degree is F!")
    else:
        print("error data")
```

寫一個程式，讓使用者輸入三邊長。
這三邊長，可能是正三角形、也可能不是正三角形，甚至無法構成三角形。

如果是正三角形，則印出如輸出範例，6 6 6正三角形的結果
如果不是三角形，則印出如輸出範例，5 4 3非正三角形的結果
若無法構成三角形，則印出，1 1 2無法構成三角形。
請輸入三組資料進行測試。

輸入範例:
6
6
6
5
4
3
1
1
2

輸出範例:
6 6 6正三角形
5 4 3非正三角形
1 1 2無法構成三角形

```python
'''
14_2.巢狀選擇-正三角形嗎?(30.0)
'''
def triangle_type(a,b,c):
     # 檢查是否可以構成三角形
    if a + b > c and a + c > b and b + c > a:
        if a == b == c:
            return f"{a} {b} {c}正三角形"
        else:
            return f"{a} {b} {c}非正三角形"
    else:
        return f"{a} {b} {c}無法構成三角形"
for _ in range(3):
    a = int(input())
    b = int(input())
    c = int(input())
    print(triangle_type(a, b, c))
```

寫一個程式來判斷你給的數，落在甚麼區域？

如果是正數，首先印出此數是正數， 小於等於 10000，則顯示 A； 其餘，則顯示 B；
在負數的情形，則印出此數是負數， 小於 -10000，則顯示 C 其餘，則顯示 D。

此程式可以執行4次。

輸入範例: 
6000
10002
-20000
-999

輸出範例: 
6000是正數
A
10002是正數
B
-20000是負數
C
-999是負數
D

```python
'''
14_3.巢狀選擇-玩玩二分法(30.0)
'''
def check_number(num):
    if num > 0 :
        print(f"{num}是正數")
        if num <= 10000 :
            print("A")
        else:
            print("B")
    if num < 0 :
        print(f"{num}是負數")
        if num <= -10000 :
            print("C")
        else:
            print("D")


# 讓使用者輸入4次資料
for _ in range(4):
    num = int(input())
    check_number(num)

#6000
#10002
#-20000
#-999    
```
