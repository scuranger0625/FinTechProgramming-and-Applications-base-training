```python
'''
17_6.重複-警示值控制-班級均分-完成(10.0)
'''


count = 0 # 歸零計數器

score_list= []

def calculation_program():
    while True:
        score = int(input())
        # 輸入-1 即結束函式
        if score == -1:
            break
        # 判分 如果區間位於0-100則印出分數
        elif score >= 0 and score <= 100 :
            
            # 宣告使用全域變數
            global score_list ,count
            score_list.append(score)
            count += 1
        # 如果輸入了-1或正確分數以外的分數
        else:
            # 判定錯誤 並重新執行
            print("error score, please input score again")
            continue
# 呼叫函式
calculation_program()


average = sum(score_list) / len(score_list)
print(f"{average:.2f}")




# 使用round() 四捨五入至指定位數
#average = round(average,2)
#print(average)

```

寫一個程式，輸入一個整數，代表分數，先不考慮不合理數據，

如果大於等於60分，則印出此分及格；
否則，印出此數不及格。
此程式可重複執行，直到分數是負數。

```python
'''
17_7.重複-警示值控制-判斷直到負分(10.0)
'''
def judgment_program():
    while True:
        score = int(input())
        if score <= 0:
            break
        elif score >= 60:
            print(f"{score}分及格")
        elif score < 60:
            print(f"{score}分不及格")

judgment_program()
    
```

```python
'''
17_8.重複-警示值控制-相反數絕對值(10.0)
'''
def judgment_program():
    while True:
        score = int(input())
        # 如果輸入為0 則跳出迴圈
        if score == 0:
            break
        # 計算相反數與絕對值
        negative_integer = -score 
        absolute_value = abs(score)
        
        # 印出結果
        print(f"{score}的相反數{negative_integer}")
        print(f"{score}的絕對值{absolute_value}")
            
judgment_program()
```

請撰寫一程式，輸入圓周長，然後計算其半徑與圓面積，並且取到小數點後2位，假設圓周率為3.14。
(程式可重複輸入，直到圓周長 <=0 為止)

輸入範例: 
16.8
68
85.66
-8

輸出範例: 
半徑2.68
面積22.47
半徑10.83
面積368.15
半徑13.64
面積584.21

```python
'''
17_9.重複-警示值控制-圓(10.0)
'''
pi = 3.14 # 假設圓周率=3.14

def method():
    while True:
        circumference = float(input()) # 圓周長
    
        if circumference <= 0:
            # 跳出迴圈 如果改成continue會出錯= =
            break
    
        else:
            # 輸入數值有效時 計算半徑和面積
            radius = circumference / (pi * 2) # 半徑
            area_circle = (radius**2) * pi # 圓面積
            print(f"半徑{radius:.2f}")
            print(f"面積{area_circle:.2f}")
        
method()
    
    
    


```

年利率公式: 利息 = 本金 * 利率 * 天數 / 365
 請利用上述公式發展一個 Python 語言程式，輸入本金、利率與天數，計算出所獲得的利息，並取小數2位
 可重複輸入，當本金為 -1 時停止輸入
輸入範例: 
1000.00
0.1
365
10000.00
0.09
1460
-1

輸出範例: 
100.00
3600.00

```python
'''
17_10.重複-警示值控制-年利率(10.0)
'''
# Interest = Principal * Rate * Days / 365

# 計算年利率公式
def calculate_interest(Principal, Rate, Days):
    Interest = Principal * Rate * Days / 365
    return round(Interest,2) # 返回小數點第2位

# 執行迴圈 直到輸入-1    
while True:
    Principal = float(input())
    
    if Principal == -1 :
        break
    
    Rate = float(input())
    Days = int(input())
    
    # 計算利息 並印出至小數第二位
    Interest = calculate_interest(Principal, Rate, Days)
    print(f"{Interest:.2f}")
    
    
    
```
