寫一個程式，輸入一個整數，代表分數，分數需在100分以內，如果大於100分，則視為不合理分數，不列入及格範圍。

如果大於等於60分，為及格
否則，為不及格
此程式可重複執行，直到分數為負數，請計算及格與不及格的人數。

輸入範例:
78
17
89
98
100
78
56
98
109
56
89
23

輸出範例: 
7個及格
4個不及格
1個不合理

```python
'''
19_1.if與迴圈的應用-及格人數(10.0)
'''

# 歸零計數器
pass_count = 0 # 及格者
fail_count = 0 # 不及格者
unreasonable_count = 0 # 不合理者


def calculate_judgement():
    while True:
        score = int(input())
        
        # 判斷為負數時 結束迴圈
        if score < 0:
            break
        
        # 全域變數發生改變 呼叫全域變數 
        global pass_count, fail_count, unreasonable_count
        
        if score > 100 :
            unreasonable_count += 1 # 不合理
            
        elif score >= 60 :
            pass_count += 1 # 及格
        
        else:
            fail_count += 1 # 不及格
            
    print(f"{pass_count}個及格")
    print(f"{fail_count}個不及格")
    print(f"{unreasonable_count}個不合理")

calculate_judgement()
```

老師想要將班級學生隨機分成三組，進行掃地區域的分配。老師想以此次的電腦成績作為分組的依據，取分數除以10，
餘數是1、2或3，為A組，
餘數是4、5或6，為B組，
餘數是7、8或9，為C組，
餘數為0者，則到他班幫忙，
超過100，則顯示錯誤輸入，
寫一個程式，此程式可以重複執行直到分數為負數，並且列出各群人數。

輸入範例: 
85
90
101
89
88
67
96
95
74
93
92
91
98
78
102
-1

輸出範例:
3個A組
4個B組
5個C組
1個他班
2個錯誤輸入

```python
'''
19_2.if與迴圈的應用-各群人數(10.0)
'''
def assign_group(score):
    if score > 100 :
        return "error"
    elif score < 0 :
        return "end" # 程式結束
    else:
        remainder = score % 10
        if remainder in[1,2,3] :
            return "A"
        elif remainder in[4,5,6] :
            return "B"
        elif remainder in[7,8,9] :
            return "C"
        elif remainder == 0 :
            return "other"
            
# 初始化各組人數計數
group_counts = {"A": 0, "B": 0, "C": 0, "other": 0, "error": 0}

# 主程式
while True:
    score= int(input())
    
    group = assign_group(score)
    
    if group == "end":
        break
    elif group in group_counts :
        # 更新對應組別的計數
        group_counts[group] += 1
    else:
        print("plz inpt again")
        
print(f"{group_counts['A']}個A組")
print(f"{group_counts['B']}個B組")
print(f"{group_counts['C']}個C組")
print(f"{group_counts['other']}個他班")
print(f"{group_counts['error']}個錯誤輸入")
        
```

請撰寫一程式，可由鍵盤輸入 x,y 座標，然後判斷該點位於哪一象限或是在座標軸上。

(程式可重複輸入，直到 x,y 座標皆為 0.0 為止)。

輸入範例: 
5.2
6.1
-2.5
-3.5
2.5
0.0
0.0
0.0

輸出範例:
第一象限
第三象限
x軸

```python
'''
19_3.if與迴圈的應用-象限(10.0)
'''
while True:
    x = float(input())
    y = float(input())
    
    # 檢查是否位於(0.0)
    if x == 0.0 and y == 0.0 :
        break
    
    # 判斷象限
    if x > 0 and y > 0 :
        print("第一象限")
    elif x < 0 and y > 0 :
        print("第二象限")
    elif x < 0 and y < 0 :
        print("第三象限")
    elif x > 0 and y < 0 :
        print("第四象限")
    elif x == 0 :
        print("y軸")
    elif y == 0 :
        print("x軸")
    
```

請撰寫一程式，給一個正整數 N，印出 1-N 之間，所有是 3 而且不是 5 的倍數以及個數。

輸入範例: 
50

輸出範例:
3 6 9 12 18 21 24 27 33 36 39 42 48
13

不換行的提示：
print(i, end=' ')

```python
'''
19_4.if與迴圈的應用-是3不是5的倍數(10.0)
'''
N = int(input())

# 歸零計數器
count = 0 

for i in range(1,N+1):
    # 判斷是3的倍數 以及不是5的倍數
    if i % 3 == 0 and i % 5 != 0:
        print(i, end=" ") # 不換行輸出
        count += 1

# 換行後輸出        
print(f"\n{count}")
```

請撰寫一程式，給一個正整數 N，找出該整數所有的正因數與個數。

輸入範例: 
24

輸出範例: 
1 2 3 4 6 8 12 24
8

```python
'''
19_5.if與迴圈的應用-因數(10.0)
'''
N = int(input())

factors= []
count = 0

for i in range(1,N+1):
    if N % i == 0 : # 判斷是否為N的因數
        factors.append(i)
        count += 1
    
# 輸出所有正因數 map每個元素應用指定的函數，並返回一個包含結果的迭代器。

print(' '.join(map(str, factors)))  # 使用空格分隔輸出

# 印出計數器
print(count)
```

計算整數1到n的偶數和，但是遇到10的倍數就終止，
分別印出偶數與總和。


輸入範例:
50

輸出範例:
2
4
6
8
20

```python
'''
19_6.if與迴圈的應用-break(15.0)
'''

n = int(input())

total_sum = 0

for i in range(1,n+1):
    if i % 2 == 0: # 判斷是否為偶數
        if i % 10 == 0 : #判斷是否為10的倍數
            break
        print(i)
        total_sum += i # 累積偶數和
        
print(total_sum)
```

寫一個程式，輸入一個整數，如果不是質數，則印出此數不是質數。

輸入範例:
78

輸出範例: 
78不是質數

```python
'''
19_7.if與迴圈的應用-不是質數(10.0)
'''
n = int(input())

if n <  2 :
    print(f"{n} 不是質數")
else:
    is_prime = True
    
    for i in range(2, int(n**0.5)+1):
        if n % i == 0 : # 如果n可以被i整除
            """
            內層迴圈檢查到某個數不是質數時，將標記變數設置為 False，
            用於跳出內層迴圈後判斷該數是否為質數。
            """
            is_prime = False
            break
            
    
    if is_prime:
        print(f"{n}是質數")
    else:
        print(f"{n}不是質數")
    
    
```
