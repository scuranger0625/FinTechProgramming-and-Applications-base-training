早餐店有三種套餐。

選擇A或a，列出牛奶、西瓜、檸檬水與吐司
選擇B或b，列出西瓜、檸檬水與吐司
選擇C或c，列出檸檬水與吐司
選擇D或d，列出吐司
其他，則是超出範圍

請寫一個程式，使得客人在選擇套餐之後，列出套餐中所含的餐點，本程式可以重複執行2次。

輸入資料:
B
F

輸出資料:
西瓜
檸檬水
吐司
超出範圍

```python
'''
15_1.多選一-點套餐(20.0)
'''
def list_meal(option):
    if option in ["A", "a"] :
        print("牛奶")
        print("西瓜")
        print("檸檬水")
        print("吐司")
    elif option in ["B","b"] :
        print("西瓜")
        print("檸檬水")
        print("吐司")
    elif option in ["C","c"] :
        print("檸檬水")
        print("吐司")
    elif option in ["D","d"] :
        print("吐司")
    else:
        print("超出範圍")

for i in range(2):
    option = input()
    list_meal(option)
```

目前台灣僅存鹽水、善化與北港有牛墟，當然牛墟已不進行牛隻買賣的交易，而是市集的型態存在。其規則如下，

日期末位數是1, 4或7, 則牛墟在鹽水
日期末位數是2, 5或8, 則牛墟在善化
日期末位數是3, 6或9, 則牛墟在北港
日期末位數是0, 則本日沒有牛墟
超過32，則是超出範圍
請輸入日期，判斷該日的牛墟狀況，本程式可以執行4次。

輸入範例:
21
20
32
15

輸出範例:
21日在鹽水
20日沒有
32日超出範圍
15日在善化

```python
'''
15_4.多選一-逛牛墟(20.0)
'''
def check_market(day):
    if day > 32 :
        print(f"{day}日超出範圍")
    else:
        # 取得日期的末位數
        last_digit = day % 10
        # 判斷牛墟的地點
        if last_digit in [1, 4, 7]:
            print(f"{day}日在鹽水")
        elif last_digit in [2 ,5 ,8]:
            print(f"{day}日在善化")
        elif last_digit in [3 ,6 ,9]:
            print(f"{day}日在北港")
        elif last_digit == 0 :
            print(f"{day}日沒有")
    
for i in range(4):
    day = int(input())
    check_market(day)
```

寫一個程式讓使用者輸入三邊長，並判斷其為正三角形、等腰三角形或直角三角形，
在可以構成三角形的狀況下，如果非以上三角形，則列為一般三角形； 若無法構成三角形，則印出無法構成三角形。

程式可執行4次。

輸入範例: 
5
6
7
2
2
5
3
4
5
3
3
4

輸出範例:
5 6 7一般三角形
2 2 5無法構成三角形
3 4 5直角三角形
3 3 4等腰三角形

```python
'''
15_5.多選一-三角形類型(20.0)
'''
def triangle_type(a, b, c):
    # 檢查是否可以構成三角形
    if a + b > c and a + c > b and b + c > a:
        # 判斷是否正為三角形
        if a == b == c :
            print(f"{a} {b} {c}正三角形")
        # 判斷是否為等腰三角形
        elif a == b or b == c or c == a :
            print(f"{a} {b} {c}等腰三角形")
        # 判斷是否為直角三角形 (符合畢氏定理)
        elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
            print(f"{a} {b} {c}直角三角形")
        else:
            print(f"{a} {b} {c}一般三角形")
    else:
            print(f"{a} {b} {c}無法構成三角形")

for i in range(4):
    a = int(input())
    b = int(input())
    c = int(input())
    triangle_type(a, b, c)
```
